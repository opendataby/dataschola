# Оценка форумных обсуждений

По мотивам публикации в газете "СБ-Беларусь сегодня" и в других источниках в связи с обсуждением на интернет-форуме портала Тут.бай дела Доната Скакуна. Само дело здесь не обсуждается.

## Опорные статьи

1. Статья Яна Жура [Когда клавиатура страшнее пистолета](https://www.sb.by/articles/kogda-klaviatura-strashnee-pistoleta.html) от 30 марта 2018 г.
2. [Реакция в Фейсбуке Юрия Зиссера на эту статью](https://www.facebook.com/yzisser/posts/10212029981124576) 31 марта 2018 г.
3. Цитата из этой реакции руководителя Talks.by Ольги Сытько: "СБ сделало аналитику комментирования отдельных статей на наших форумах talks.by. Результат они считают достоверным - якобы обрабатывали программой, выводы сделали сами. Штука в том, что двух из трёх якобы оплаченных комментаторов я ЗНАЮ ЛИЧНО И ГОТОВА ПОДПИСАТЬСЯ, что они искренне пишут то, что думают. СБ, ваши выводы смехотворны. Зачем вы берётесь анализировать данные в области, в которой ничего не смыслите? В следующий раз позовите меня. )"

## Поля в массивах с данными обсуждений:

1. **user_name** - имя комментатора
2. **user_id** - id комментатора
3. **comment_time** - время публикации комментария
4. **comment_country** - страна комментатора
5. **comment_body** - текст комментария
6. **quote_time** - время публикации процитированного комментария (если комментарий содержит цитату)
7. **quoted_user** - процитированный пользователь (если комментарий содержит цитату)
8. **likes** - число лайков у комментария
9. **dislikes** - число дислайков у комментария

## Поля в массиве с данными пользователей:

1. **user_id** - id комментатора
2. **total_comments** - всего комментариев
3. **date_registered** - дата регистрации на форуме

## Подводные камни

1. Общее количество оставленных комментариев, указанное на странице пользователя, может не соответствовать действительности. Например, на странице пользователя [Денис_Барковский](https://talks.by/member.php?u=753053.html) указан 0 комментариев (по состоянию на 1 апреля 2018), но если там же кликнуть на ссылку [Найти все сообщения](https://talks.by/search.php?do=finduser&userid=753053&contenttype=vBForum_Post&showposts=1), то их окажется больше сотни.

2. Непонятка с количеством комментариев. На одной странице обсуждения выводится 50 комментариев. В первом 208 страниц, т.е. ожидается около 10 400 комментариев - по факту удалось собрать всего 9 939. Во втором обсуждении обсуждении 159 страниц, т.е. ожидается порядка 7950 комментариев - по факту собрано 7 799.
